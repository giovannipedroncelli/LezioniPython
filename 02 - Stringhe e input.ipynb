# Stringhe
Le stringhe in Python sono le variabili di tipo str. I valori delle stringhe si possono inserire tra doppi apici " " o tra apici singoli ' '.

````python
mia_stringa = 'questa è una stringa!'
mia_stringa2 = "anche questa è una stringa!!!"
````

Se la stringa deve contenere un apice doppio o un apice singolo bisogna includere \ nella stringa per non incorrere in errori:

````python
stringa = 'L\'auto è nel garage.'
print(stringa)
````

Se ci dimentichiamo del carattere \ avremo un errore:

````python
stringa = 'L'auto è nel garage.'
````

In Python si possono effettuare alcune operazioni aritmetiche con le stringhe

````python
prima_parola = 'Ciao'
seconda_parola = 'amico'
````

````python
print(prima_parola + seconda_parola)
````

````python
print(prima_parola + ' ' + seconda_parola)
````

````python
print(prima_parola * 5)
````

Possiamo anche stampare i singoli caratteri della stringa usando gli indici: l'indice del primo carattere è sempre 0.

````python
print(prima_parola[0])
````

````python
print(prima_parola[1])
````

Python supporta anche gli indici negativi:

````python
print(prima_parola[-1])
````

````python
print(prima_parola[-2])
````

### La funzione len()

`len()` è una funzione di Python che restituisce la lunghezza di un oggetto come una stringa. La lunghezza di una stringa è il numero di caratteri che essa contiene ed è sempre un intero.

````python
print(len(prima_parola))
````

````python
print(len("ababa") / len("ab"))
````

### Il metodo format()
Si tratta di un metodo molto utile all'interno delle funzioni `print()`.

Il metodo `format()` permette di inserire all'interno di una stringa dei valori di variabili.

````python
print("Giovanni ha {} anni".format(37))
````

````python
animale = "cane"
azione = "morde"
print("Il tuo {} {}?".format(animale, azione))
````

````python
animale = "canarino"
azione = "canta"
print("Il tuo {} {}?".format(animale, azione))
````

````python
stringa = "Maria studia {} e {}"
print(stringa.format("informatica", "sistemi"))
````

Possiamo anche formattare le stringhe con una sintassi più sintetica, utilizzando le cosiddette *f-string*.

Le *f-string* sono stringhe precedute dal carattere `f` e contengono delle espressioni tra parentesi graffe `{}`.

Le espressioni possono essere delle variabili, delle operazioni, delle chiamate a funzioni, ecc.

Le *f-string* vengono valutate e il loro risultato viene inserito nella stringa.

````python
nome = "Mario"
cognome = "Rossi"
print(f"Mi chiamo {nome} {cognome}")
````

### Input

Per poter chiedere all'utente del programma un valore in input, in Python si usa la funzione ```input()```.

````python
nome = input("Inserisci il tuo nome")
````

Questa riga di codice mostra all'utente il messaggio "Inserisci il tuo nome" e salva il valore che viene inserito dall'utente all'interno della variabile chiamata nome

**Tutti i valori inseriti dall'utente sono SEMPRE di tipo STRINGA**

Esempio di codice con output inatteso

````python
numero = input("Inserisci un numero e io stamperò il suo doppio")
print(numero * 2)
````

Se l'utente inserisce in input il valore "23", il programma stamperà il valore "2323" perché la variabile numero è di tipo stringa e non intero.

Per far funzionare adeguatamente questo piccolo programma è necessario convertire l'input dell'utente (**UNA STRINGA**) nel tipo di dato desiderato.

In Python ci sono 4 funzioni che permettono di convertire il tipo di un valore o di una variabile:

1. `int()` per convertire un valore in un intero
2. `float()` per convertire un valore in un numero con la virgola
3. `bool()` per convertire un valore in un booleano (True cioè 1 oppure False cioè 0)
4. `str()` per convertire un valore in una stringa

Possiamo quindi riscrivere il programma precedente in questo modo:

````python
numero = int(input("Inserisci un numero e io stamperò il suo doppio"))
print(numero * 2)
````

Adesso, se l'utente inserisce il valore "23", prima di essere salvato nella variabile numero, verrà convertito in un intero, grazie all'applicazione della funzione int(). Quindi, il programma stamperà il valore 46.
